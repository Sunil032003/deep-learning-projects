```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Input,Dropout
```

```python
df=pd.read_csv('/content/student-mat.csv',sep=';')
```

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
```

```python
le=OneHotEncoder()
```

```python
catergorical=df.select_dtypes(include='object').columns
```

```python
print(catergorical)
```

```python
df=pd.get_dummies(df,columns=catergorical,drop_first=True)
```

```python
# Convert boolean columns (from one-hot encoding) to integers (0 or 1)
for col in df.select_dtypes(include='bool').columns:
    df[col] = df[col].astype(int)

print("DataFrame after converting boolean columns to int:")
display(df.head())
```

```python
df.head()
```

```python
x=df.drop('G3',axis=1)
y=df['G3']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
model=Sequential([
Input(shape=(41,)),
Dense(128,activation='relu'),
Dense(64,activation='relu'),
Dense(32,activation='relu'),
Dense(1,activation='relu')])
```

```python
model.compile(optimizer='adam',loss='mse',metrics=['r2_score'])
```

```python
model.fit(x_train,y_train,epochs=500,batch_size=32,validation_data=(x_test,y_test))
```

```python
predict=model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
print(r2_score(y_test,predict))
```
